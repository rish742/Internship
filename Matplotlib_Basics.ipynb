# Line plot or Line chart in Python with Legends

```python
# Create simple Line chart in Python

import matplotlib.pyplot as plt

# %matplotlib inline

values = [1, 5, 8, 9, 7, 11, 8, 12, 14, 9]

plt.plot(values)

plt.show()
```

Line 3: Imports the pyplot function of matplotlib library in the name of plt.

Line 5: It is a magic function that renders the figure in a notebook
    
Line 7: Inputs the array to the variable named values
    
Line 8: Plots the line chart with values and choses the x axis range from 1 to 11.
    
Line 9: Displays the resultant line chart in python'''

```python
# Multiple Line chart in Python with legends and Labels

sales1 = [1, 5, 8, 9, 7, 11, 8, 12, 14, 9, 5]

sales2 = [3, 7, 9, 6, 4, 5, 14, 7, 6, 16, 12]

line_chart1 = plt.plot(range(1, 12), sales1)

line_chart2 = plt.plot(range(1, 12), sales2)

plt.title('Monthly sales of 2016 and 2017')

plt.xlabel('Sales')

plt.ylabel('Month')

plt.legend(['year 2016', 'year 2017'], loc=4)

plt.show()
```

lets take an example of sale of units in 2016 and 2017 to demonstrate line chart in python.

Line 3 and 5: Inputs the arrays to the variables named sales1 and sales2.

Line 7 and 9: Plots the line charts (line_chart1 and line_chart2) with sales1 and sales 2 

and choses the x axis range from 1 to 12.

Line 11: Gets the title for the plot

Line 13 and 15: Gets the label for x and y axis respectively

Line 17: plots the legend for line_chart1 and line_chart2.

Line 19: Displays the resultant multiple line chart'''


Line styles for python line chart:
    
| Character | Line Style |
| --- | --- |
| ‘‐’ | Solid line |
| ‘‐‐’ | Dashed line |
| ‘‐.’ | Dash‐dot line |
| ‘:’ | Dotted line |

```python
# Multiple line chart in python with different line style

sales1 = [1, 5, 8, 9, 7, 11, 8, 12, 14, 9, 5]

sales2 = [3, 7, 9, 6, 4, 5, 14, 7, 6, 16, 12]

line_chart1 = plt.plot(range(1, 12), sales1, '--')

line_chart2 = plt.plot(range(1, 12), sales2, ':')

plt.title('Monthly sales of 2016 and 2017')

plt.xlabel('Sales')

plt.ylabel('Month')

plt.legend(['year 2016', 'year 2017'], loc=4)

plt.show()
```

```python
# Multiple line chart in python with specified line color

sales1 = [1, 5, 8, 9, 7, 11, 8, 12, 14, 9, 5]

sales2 = [3, 7, 9, 6, 4, 5, 14, 7, 6, 16, 12]

line_chart1 = plt.plot(range(1, 12), sales1, 'g')

line_chart2 = plt.plot(range(1, 12), sales2, 'm')

plt.title('Monthly sales of 2016 and 2017')

plt.xlabel('Sales')

plt.ylabel('Month')

plt.legend(['year 2016', 'year 2017'], loc=4)

plt.show()
```

# Pie Chart in Python with Legends

```python
values = [60, 80, 90, 55, 10, 30]

colors = ['b', 'g', 'r', 'c', 'm', 'y']

labels = ['US', 'UK', 'India', 'Germany', 'Australia', 'South Korea']

explode = (0.2, 0, 0, 0, 0, 0)

plt.pie(values, colors=colors, labels=values,
        explode=explode, counterclock=False, shadow=True)

plt.title('Population Density Index')

plt.legend(labels, loc=3)

plt.show()
```

Line 1: Inputs the arrays to the variables named values which denotes happiness index in our case.

Line 3 and Line 5: Array of input colors and array of labels which are countries

Line 7: first value is exploded out (projected out) by 0.2

Line 9: inputs all above values to pie() function of pyplot. Values are displayed clock wise with counterclock=False.

Line 11: Assigns Title to the pie chart.

Line 13 and Line 15: adds Legend and places at location 3 which is bottom left corner and Shows the pie chart with legend.

```python
# Create Pie chart in Python with percentage values

values = [60, 80, 90, 55, 10, 30]

colors = ['b', 'g', 'r', 'c', 'm', 'y']

labels = ['US', 'UK', 'India', 'Germany', 'Australia', 'South Korea']

explode = (0.2, 0, 0, 0, 0, 0)

plt.pie(values, colors=colors, labels=labels,

        explode=explode, autopct='%1.1f%%',

        counterclock=False, shadow=True)

plt.title('Population Density Index')

plt.show()
```

Line 19: Inputs all above values, colors, label to pie() function of pyplot. 

Argument named autopct converts the values in terms of percentages and plots it in the pie chart. 

Values are displayed clock wise with counterclock=False.

# Scatter plot in Python using matplotlib

```python
weight1 = [63.3, 57, 64.3, 63, 71, 61.8, 62.9,
           65.6, 64.8, 63.1, 68.3, 69.7, 65.4, 66.3, 60.7]

height1 = [156.3, 100.7, 114.8, 156.3, 237.1, 123.9, 151.8,
           164.7, 105.4, 136.1, 175.2, 137.4, 164.2, 151, 124.3]

plt.scatter(weight1, height1, c='b', marker='o')

plt.xlabel('weight', fontsize=16)

plt.ylabel('height', fontsize=16)

plt.title('scatter plot - height vs weight', fontsize=20)

plt.show()
```

Line 1 and Line 3: Inputs the arrays to the variables named weight1 and height1.

Line 5: scatter function which takes takes x axis (weight1) as first argument, y axis (height1) as second argument, 

colour is chosen as blue in third argument and marker=’o’ denotes the type of plot, Which is dot in our case.

Line 7 and Line 9: x label and y label with desired font size is created.

Line 11 and Line 13: Mentions the Chart Title with font size and scatter plot is shown.

Different values for markers and their representation is shown below.

| Marker | meaning |
| --- | --- |
| . | point |
| , | pixel |
| o | circle |
| v | triangle_down |
| ^ | triangle_up |
| 8 | octagon |
| s | sqaure |
| p | pentagon |
| * | star |
| h | hexagon |	    
| + | plus |	    
| D | diamond |

```python
# Create Scatter plot by Groups in Python

import numpy as np

weight1 = [57, 58.2, 58.6, 59.6, 59.8, 60.2, 60.5,
           60.6, 60.7, 61.3, 61.3, 61.4, 61.8, 61.9, 62.3]

height1 = [100.7, 195.6, 94.3, 127.1, 111.7, 159.7, 135,
           149.9, 124.3, 112.9, 176.7, 110.2, 123.9, 161.9, 107.8]

weight2 = [62.9, 63, 63.1, 63.2, 63.3, 63.4, 63.4, 63.4,
           63.5, 63.6, 63.7, 64.1, 64.3, 64.3, 64.7, 64.8, 65]

height2 = [151.8, 156.3, 136.1, 124.2, 156.3, 130, 181.2, 255.9,
           163.1, 123.1, 119.5, 179.9, 114.8, 174.1, 108.8, 105.4, 141.4]

weight3 = [69.2, 69.2, 69.4, 69.7, 70, 70.3, 70.8,
           71, 71.1, 71.7, 71.9, 72.4, 73, 73.1, 76.2]

height3 = [166.8, 172.9, 193.8, 137.4, 162.4, 137.1, 169.1,
           237.1, 189.1, 179.3, 174.8, 213.3, 198, 191.1, 220.6]

weight = np.concatenate((weight1, weight2, weight3))

height = np.concatenate((height1, height2, height3))

color_array = ['b'] * 15 + ['g'] * 17 + ['r']*15

plt.scatter(weight, height, marker='*', c=color_array)

plt.xlabel('weight', fontsize=16)

plt.ylabel('height', fontsize=16)

plt.title('grouped scatter plot - height vs weight', fontsize=20)

plt.show()
```

weight1, weight2 and weight3 is concatenated to form a weight array.

height1, height2 and height3 is concatenated to form a height array.

Corlor_array with first 15 blue color, second 15 green color and last 15 red color is created.

All the above three arguments along with the marker=’*’ is passed to Scatter function.

Which results in plotting the scatter plot of 3 different groups with 3 different colors with ‘*’ being used as plot.

# Bar Plot or Bar Chart in Python with legend

```python
city = ['Delhi', 'Beijing', 'Washington', 'Tokyo', 'Moscow']

pos = np.arange(len(city))

Happiness_Index = [60, 40, 70, 65, 85]

plt.bar(pos, Happiness_Index, color='blue', edgecolor='black')

plt.xticks(pos, city)

plt.xlabel('City', fontsize=16)

plt.ylabel('Happiness_Index', fontsize=16)

plt.title('Barchart - Happiness index across cities', fontsize=20)

plt.show()
```

Line number 3, arange() function sets the x axis value.

Line number 7, bar() function takes both the axis as input, sets color as blue and border color as black.

Line number 9, xticks() function takes value as labels i.e. label of cities and happiness_index as input and plots the label

Line number 11 to 13, plots x label, y label and title and shows the output.

```python
# Horizontal Bar Chart in Python

city = ['Delhi', 'Beijing', 'Washington', 'Tokyo', 'Moscow']

pos = np.arange(len(city))

Happiness_Index = [60, 40, 70, 65, 85]

plt.barh(pos, Happiness_Index, color='blue', edgecolor='black')

plt.yticks(pos, city)

plt.xlabel('Happiness_Index', fontsize=16)

plt.ylabel('City', fontsize=16)

plt.title('Barchart - Happiness index across cities', fontsize=20)

plt.show()
```

Line number 9, barh() function plots the horizontal bar chart which takes both the axis as input, sets color as blue and border color as black.

```python
# Stacked Bar Chart in Python with legends

city = ['Delhi', 'Beijing', 'Washington', 'Tokyo', 'Moscow']

Gender = ['Male', 'Female']

pos = np.arange(len(city))

Happiness_Index_Male = [60, 40, 70, 65, 85]

Happiness_Index_Female = [30, 60, 70, 55, 75]

plt.bar(pos, Happiness_Index_Male, color='blue', edgecolor='black')

plt.bar(pos, Happiness_Index_Female, color='pink',
        edgecolor='black', bottom=Happiness_Index_Male)

plt.xticks(pos, city)

plt.xlabel('City', fontsize=16)

plt.ylabel('Happiness_Index', fontsize=16)

plt.title('Stacked Barchart - Happiness index across cities', fontsize=18)

plt.legend(Gender, loc=2)

plt.show()
```

Line number 13, bar() functions plots the Happiness_Index_Male first.

Line number 15, bar() function plots the Happiness_Index_Female on top of Happiness_Index_Male with the help of argument bottom=Happiness_Index_Male.

Legend is plotted on the top left corner.

```python
# Grouped Bar Chart in Python with legends

city = ['Delhi', 'Beijing', 'Washington', 'Tokyo', 'Moscow']

Gender = ['Male', 'Female']

pos = np.arange(len(city))

bar_width = 0.35

Happiness_Index_Male = [60, 40, 70, 65, 85]

Happiness_Index_Female = [30, 60, 70, 55, 75]

plt.bar(pos, Happiness_Index_Male, bar_width, color='blue', edgecolor='black')

plt.bar(pos+bar_width, Happiness_Index_Female,
        bar_width, color='pink', edgecolor='black')

plt.xticks(pos, city)

plt.xlabel('City', fontsize=16)

plt.ylabel('Happiness_Index', fontsize=16)

plt.title('Group Barchart - Happiness index across cities By Gender', fontsize=18)

plt.legend(Gender, loc=2)

plt.show()
```

Line number 9, assigns bar width is fixed to 0.35.

Line number 15, bar() functions plots the Happiness_Index_Male first.

Line number 17, bar() function plots the Happiness_Index_Female side wise of Happiness_Index_Male through the first argument pos+bar_width.

Legend is plotted on the top left corner.

# Box plot in Python with matplotlib

```python
# Four arrays are passed to an array named box_plot_data and it is plotted with boxplot function. Which is shown below.

value1 = [82, 76, 24, 40, 67, 62, 75, 78, 71,
          32, 98, 89, 78, 67, 72, 82, 87, 66, 56, 52]

value2 = [62, 5, 91, 25, 36, 32, 96, 95, 3, 90,
          95, 32, 27, 55, 100, 15, 71, 11, 37, 21]

value3 = [23, 89, 12, 78, 72, 89, 25, 69, 68,
          86, 19, 49, 15, 16, 16, 75, 65, 31, 25, 52]

value4 = [59, 73, 70, 16, 81, 61, 88, 98, 10,
          87, 29, 72, 16, 23, 72, 88, 78, 99, 75, 30]

box_plot_data = [value1, value2, value3, value4]

plt.boxplot(box_plot_data)

plt.show()
```

```python
# Create box plot in python with fills and labels

value1 = [82, 76, 24, 40, 67, 62, 75, 78, 71,
          32, 98, 89, 78, 67, 72, 82, 87, 66, 56, 52]

value2 = [62, 5, 91, 25, 36, 32, 96, 95, 3, 90,
          95, 32, 27, 55, 100, 15, 71, 11, 37, 21]

value3 = [23, 89, 12, 78, 72, 89, 25, 69, 68,
          86, 19, 49, 15, 16, 16, 75, 65, 31, 25, 52]

value4 = [59, 73, 70, 16, 81, 61, 88, 98, 10,
          87, 29, 72, 16, 23, 72, 88, 78, 99, 75, 30]

box_plot_data = [value1, value2, value3, value4]

plt.boxplot(box_plot_data, patch_artist=True, labels=[
            'course1', 'course2', 'course3', 'course4'])

plt.show()
```

boxplot() function takes the data array to be plotted as input in first argument, second argument patch_artist=True , fills the boxplot and third argument takes the label to be plotted.

```python
# Create box plot in python with notch

value1 = [82, 76, 24, 40, 67, 62, 75, 78, 71,
          32, 98, 89, 78, 67, 72, 82, 87, 66, 56, 52]

value2 = [62, 5, 91, 25, 36, 32, 96, 95, 3, 90,
          95, 32, 27, 55, 100, 15, 71, 11, 37, 21]

value3 = [23, 89, 12, 78, 72, 89, 25, 69, 68,
          86, 19, 49, 15, 16, 16, 75, 65, 31, 25, 52]

value4 = [59, 73, 70, 16, 81, 61, 88, 98, 10,
          87, 29, 72, 16, 23, 72, 88, 78, 99, 75, 30]

box_plot_data = [value1, value2, value3, value4]

plt.boxplot(box_plot_data, notch='True', patch_artist=True,
            labels=['course1', 'course2', 'course3', 'course4'])

plt.show()
```

boxplot() function takes the data array to be plotted as input in first argument, second argument notch=‘True’ creates the notch format of the box plot. Third argument patch_artist=True, fills the boxplot with color and fourth argument takes the label to be plotted.

```python
# Horizontal box plot in python with different colors

value1 = [82, 76, 24, 40, 67, 62, 75, 78, 71,
          32, 98, 89, 78, 67, 72, 82, 87, 66, 56, 52]

value2 = [62, 5, 91, 25, 36, 32, 96, 95, 3, 90,
          95, 32, 27, 55, 100, 15, 71, 11, 37, 21]

value3 = [23, 89, 12, 78, 72, 89, 25, 69, 68,
          86, 19, 49, 15, 16, 16, 75, 65, 31, 25, 52]

value4 = [59, 73, 70, 16, 81, 61, 88, 98, 10,
          87, 29, 72, 16, 23, 72, 88, 78, 99, 75, 30]

box_plot_data = [value1, value2, value3, value4]

box = plt.boxplot(box_plot_data, vert=0, patch_artist=True, labels=['course1', 'course2', 'course3', 'course4'],
                  )

colors = ['cyan', 'lightblue', 'lightgreen', 'tan']

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

boxplot() function takes argument vert =0 which plots the horizontal box plot.
Colors array takes up four different colors and passed to four different boxes of the boxplot with patch.set_facecolor() function.

# Create Histogram in Python using matplotlib

```python
values = [82, 76, 24, 40, 67, 62, 75, 78, 71,
          32, 98, 89, 78, 67, 72, 82, 87, 66, 56, 52]

plt.hist(values, 5, histtype='step', align='mid',
         color='g', label='Test Score Data')

plt.legend(loc=2)

plt.title('Histogram of score')

plt.show()
```

Line 1: Inputs the arrays to the variables named values which is score in our case.

Line 3: In hist function, first argument accepts the values to be plotted, second argument is the number of bins, histype=’step’ which plots the histogram in step format, aligned to mid, color chosen is green.

Line 5: Legend function argument loc=2 plots the legend on the top left corner.

Line 7 and Line 9: Assign Title to Histogram and Histogram is plotted.

```python
# Create Histogram in Python with bar Filled

values = [82, 76, 24, 40, 67, 62, 75, 78, 71,
          32, 98, 89, 78, 67, 72, 82, 87, 66, 56, 52]

plt.hist(values, 5, histtype='bar', align='mid', color='c',
         label='Test score Data', edgecolor='black')

plt.legend()

plt.title('Histogram of score')

plt.show()
```

Line 5: In hist function, first argument accepts the values to be plotted, second argument is the number of bins, histype=’bar’ which plots the histogram in bar filled format, aligned to mid, color chosen is cyan. edgecolor=’black’ indicates the black borders for histogram.
